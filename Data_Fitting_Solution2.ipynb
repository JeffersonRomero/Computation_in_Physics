```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

n = 6               # Grado del polinomio
a=np.zeros((n+1,))

x = np.linspace(-1,1,101)
y = np.zeros_like(x)
sigma = (2*np.random.rand(len(x))-1)/50

for i in range(n+1):
    a[i] = rd.uniform(-1,1)
    y += a[i]*x**i

y += sigma

p = np.polyfit(x,y,n)
poly = np.poly1d(p)(x)

plt.plot(x,y)
plt.plot(x,poly)
plt.show()

print(a)
print(np.flip(p))

f = np.transpose([x,y])
np.savetxt('Polynomial_Data2.dat',f)
```

```python
dfx=np.array(df[0]) #Agregamos los valores de la columna x en formato array a una variable
```

```python
dfx #Array de los valores de x
```

```python
#Calculamos las columnas con el número de polinomios (n=6) del array x.
df0=dfx**0
df1=dfx**1
df2=dfx**2
df3=dfx**3
df4=dfx**4
df5=dfx**5
df6=dfx**6
```

```python
matrix_x = np.column_stack((df0, df1, df2, df3, df4, df5, df6)) #Creamos la matrix de los valores de x para cada polinomio
```

```python
matrix_x #Matriz de x con 7 columnas dado 6 polinomios
```

```python
matrix_y=np.array(df[1]) #Creamos el array para los valores de y
```

```python
matrix_y #Muestra la columna de los valores de y (matriz Y)
```

```python
matrix_xT=matrix_x.T #Calculamos la transpuesta de la matriz de x
```

```python
matrix_xT #Muestra la transpuesta de la matriz x.
```

```python
matrix_squarex=np.dot(matrix_xT,matrix_x) 
#hacemos el producto matricial de la matriz transpuesta de x y la matriz x para calcular la matriz cuadrada
```

```python
matrix_squarex #Muestra la matrix cuadrada de X
```

```python
matrix_squarex_inverse=np.linalg.inv(matrix_squarex)
#Calculamos la matrix inversa de x haciendo uso de la matriz cuadra obtenida de x
```

```python
matrix_squarex_inverse #Muestra la matriz cuadrada inversa de x
```

```python
matrix_xT_y=np.dot(matrix_xT,matrix_y) #Multiplicamos la transpuesta de x por la matriz y.
```

```python
matrix_xT_y #Muestra el producto matricial de la transpuesta de x por la matriz y.
```

```python
coeficients_a=np.dot(matrix_squarex_inverse,matrix_xT_y)
#Multiplicamos la matriz cuadrada inversa de x por la matriz obtenida del 
#producto matricial de la transpuesta de x y la matriz y
```

```python
coeficients_a #Estos son los coeficentes de a que queríamos.
```
