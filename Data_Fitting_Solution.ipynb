# Fórmula general de los mínimos cuadrados en forma matricial:
# $$ \mathbf{Y} = \mathbf{A} \mathbf{X}  $$
# Al multiplicar a ambos lados por la transpuesta de X:
# $$ \mathbf{X}^T\mathbf{Y} = \mathbf{X}^T\mathbf{X} \mathbf{A} $$
# Al despejar A (que nos da los coeficientes que vamos a calcular): 
# $$ \mathbf{A}=(\mathbf{X}^T \mathbf{X}) ^{-1} (\mathbf{X}^T\mathbf{Y}) $$

```python
import numpy as np #Importamos numpy para trabajar con matrices
```

```python
dfx=np.array(df[0]) #Agregamos los valores de la columna x en formato array a una varia
```

```python
print(f'los valores del array son: \n\n{dfx}') #Array de los valores de x
```

```python
#Calculamos las columnas con el número de polinomios (n=6) del array x.
df0=dfx**0
df1=dfx**1
df2=dfx**2
df3=dfx**3
df4=dfx**4
df5=dfx**5
df6=dfx**6
```

```python
matrix_x = np.column_stack((df0, df1, df2, df3, df4, df5, df6)) #Creamos la matrix de los valores de x para cada polinomio
```

# $$\mathbf{X}=$$

```python
print(f'La matriz X con un grado polinomial de 6 es: \n\n{matrix_x}') #Matriz de x con 7 columnas dado 6 polinomios
```

```python
dfy=np.array(df[1]) #Creamos el array para los valores de y
```

# $$\mathbf{Y}=$$

```python
print(f'la matriz Y es: \n\n{dfy}') #Muestra la columna de los valores de y
```

```python
matrix_xT=matrix_x.T #Calculamos la transpuesta de la matriz de x
```

# $$\mathbf{X^T}=$$

```python
print(f'La matriz transpuesta de X es: \n\n{matrix_xT}') #Muestra la transpuesta de la matriz x
```

```python
matrix_squarex=np.dot(matrix_xT,matrix_x) 
#hacemos el producto matricial de la matriz transpuesta de x y la matriz x para calcular la matriz cuadrada
```

# $$\mathbf{X^T X}=$$

```python
print(f'La matriz cuadrada es: \n\n{matrix_squarex}') #Muestra la matrix cuadrada de X
```

```python
matrix_squarex_inverse=np.linalg.inv(matrix_squarex)
#Calculamos la matrix inversa de x haciendo uso de la matriz cuadra obtenida de x
```

# $$\mathbf{(X^T X)^{-1}}=$$

```python
print(f'La matriz cuadrada e inversa de X es: \n\n{matrix_squarex_inverse}') #Muestra la matriz cuadrada inversa de x
```

```python
matrix_xT_y=np.dot(matrix_xT, dfy) #Multiplicamos la transpuesta de x por la matriz y.
```

# $$(\mathbf{X^T}\mathbf{Y})=$$

```python
print(f'El producto matricial de la transpuesta de X por Y es: \n\n{matrix_xT_y}') #Muestra el producto matricial de la transpuesta de x por la matriz y.
```

```python
a=np.dot(matrix_squarex_inverse,matrix_xT_y)
#Multiplicamos la matriz cuadrada inversa de x por la matriz obtenida del 
#producto matricial de la transpuesta de x y la matriz y
```

# $$\mathbf{A}=$$

```python
print(f'\nLos coeficientes a obtenidos después del procedimiento usando matrices son: \n\n{a}') #Coeficientes calculados usando matrices
```
